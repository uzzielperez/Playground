```python
import ROOT
from ROOT import TCanvas, TPad, TFile, TPaveText, TColor, TGaxis, TH1F, TPad, TH1D, TLegend
from ROOT import kBlack, kBlue, kRed
from ROOT import gBenchmark, gStyle, gROOT, gDirectory
```

```python
path = "/uscms_data/d3/cuperez/CMSSW_8_0_25/src/scripts/macros/DiphotonAnalysis/"
filename = "diphoton16bkg_histograms.root"
```

```python
f = ROOT.TFile.Open(path+filename)
f.ls()
```

```python
%jsroot on
c1=ROOT.TCanvas("mycanvas","mycanvas",600,600)
#c1.SetGrid( )

diphoton_Minv = f.Get("diphotonMinv")
diphoton_Minv.GetXaxis().SetTitleOffset(1.4)
diphoton_Minv.GetXaxis().SetTitle(r"m_{#gamma#gamma} scale[0.8]{(GeV)}")
diphoton_Minv.GetYaxis().SetTitleOffset(1.4)
diphoton_Minv.GetYaxis().SetTitle("weighted events/ 80 GeV")
diphoton_Minv.SetAxisRange(200, 1600) #To draw hist from 200 1600
diphoton_Minv.Draw("hist") #have to add hist to not draw with error bars

leg1 = TLegend(.60, 0.75, .97, .85)
leg1.SetBorderSize(0)
leg1.SetFillColor(0)
leg1.SetFillStyle(0)
leg1.SetTextFont(42)
leg1.SetTextSize(0.035)
leg1.AddEntry(diphoton_Minv,"SM Diphoton","l")
leg1.Draw()

#diphoton_Minv.SetLineStyle(1) #Default

c1.SetLogy()
#c1.SetFillStyle(3001)
#c1.SetFillColor(kRed)
c1.Draw()
```

```python
filename2="diphoton2016bkg_histograms.root"
f2 = ROOT.TFile.Open(path+filename2)
#f2.ls()
```

```python
%jsroot on
c=ROOT.TCanvas("canvas","mycanvas",600,600)
diphoton_Minvn = f2.Get("diphotonMinvn")
diphoton_Minvn.GetXaxis().SetTitleOffset(1.4)
diphoton_Minvn.GetXaxis().SetTitle(r"m_{#gamma#gamma} scale[0.8]{(GeV)}")
diphoton_Minvn.GetYaxis().SetTitleOffset(1.4)
diphoton_Minvn.GetYaxis().SetTitle("weighted events/ 80 GeV")
diphoton_Minvn.SetAxisRange(200, 2000) #To draw hist from 200 1600
diphoton_Minvn.Draw("hist") #have to add hist to not draw with error bars

legend1 = TLegend(.60, 0.75, .97, .85)
legend1.SetBorderSize(0)
legend1.SetFillColor(0)
legend1.SetFillStyle(0)
legend1.SetTextFont(42)
legend1.SetTextSize(0.035)
legend1.AddEntry(diphoton_Minvn,"SM Diphoton","l")
legend1.Draw()


c.SetLogy()
c.Draw()
```

```python
canvas = ROOT.TCanvas()
canvas.Clear()


hphoton1pTn = f2.Get("photon1Ptn")
hphoton2pTn = f2.Get("photon2Ptn")

hphoton1pTn.GetXaxis().SetTitleOffset(1.4)
hphoton1pTn.GetXaxis().SetTitle(r"p_{T} (GeV)")
hphoton1pTn.SetAxisRange(200, 2000)
hphoton1pTn.GetYaxis().SetTitleOffset(1)
hphoton1pTn.GetYaxis().SetTitle("weighted events/ 80 GeV")

hphoton1pTn.Draw("hist")
hphoton2pTn.SetLineColor(kBlack)
hphoton2pTn.Draw("hist same")


#diphoton_Minvn.GetXaxis().SetTitleOffset(1.4)
#diphoton_Minvn.GetXaxis().SetTitle(r"m_{#gamma#gamma} scale[0.8]{(GeV)}")
#diphoton_Minvn.GetYaxis().SetTitleOffset(1.4)
#diphoton_Minvn.GetYaxis().SetTitle("weighted events/ 80 GeV")
#diphoton_Minvn.SetAxisRange(200, 2000) #To draw hist from 200 1600

#leg = TLegend(.73,0.32,.97,.53)
leg = TLegend(.75, 0.75, .97, .85)
leg.SetBorderSize(0)
leg.SetFillColor(0)
leg.SetFillStyle(0)
leg.SetTextFont(42)
leg.SetTextSize(0.035)
leg.AddEntry(hphoton1pTn,r"#gamma_{1} p_{T}")
leg.AddEntry(hphoton2pTn,r"#gamma_{2} p_{T}")
leg.Draw()

canvas.SetLogy()
canvas.Draw()

#canvas.cd(1); diphoton_Minvn.Draw("hist")
#canvas.cd(2); diphoton_Minvn.Draw("hist")
#canvas.cd(3); diphoton_Minvn.Draw("hist")
#canvas.cd(3); diphoton_Minvn.Draw("hist")
```

```python
mcanvas = ROOT.TCanvas()
mcanvas.Clear()
mcanvas.Divide(2,1)

hphoton1etan = f2.Get("photon1Etan")
hphoton1phin = f2.Get("photon1Phin")
hphoton2etan = f2.Get("photon2Etan")
hphoton2phin = f2.Get("photon2Phin")

mcanvas.cd(1)

hphoton1etan.GetXaxis().SetTitleOffset(1)
hphoton1etan.GetXaxis().SetTitle(r"#eta")
hphoton1etan.SetAxisRange(200, 2000)
hphoton1etan.GetYaxis().SetTitleOffset(1.2)
hphoton1etan.GetYaxis().SetTitle("weighted events/ 80 GeV")

hphoton1etan.Draw("hist")
hphoton2etan.SetLineColor(kBlack)
hphoton2etan.Draw("hist same")

leg2 = TLegend(.68, 0.75, .97, .85)
leg2.SetBorderSize(0)
leg2.SetFillColor(0)
leg2.SetFillStyle(0)
leg2.SetTextFont(42)
leg2.SetTextSize(0.035)
leg2.AddEntry(hphoton1etan,r"#gamma_{1} p_{T}")
leg2.AddEntry(hphoton2etan,r"#gamma_{2} p_{T}")
leg2.Draw()

mcanvas.cd(2)

hphoton1phin.GetXaxis().SetTitleOffset(1)
hphoton1phin.GetXaxis().SetTitle(r"#phi")
hphoton1phin.SetAxisRange(200, 2000)
hphoton1phin.GetYaxis().SetTitleOffset(1.2)
hphoton1phin.GetYaxis().SetTitle("weighted events/ 80 GeV")

hphoton1phin.Draw("hist")
hphoton2phin.SetLineColor(kBlack)
hphoton2phin.Draw("hist same")

leg3 = TLegend(.68, 0.75, .97, .85)
leg3.SetBorderSize(0)
leg3.SetFillColor(0)
leg3.SetFillStyle(0)
leg3.SetTextFont(42)
leg3.SetTextSize(0.035)
leg3.AddEntry(hphoton1phin,r"#gamma_{1}")
leg3.AddEntry(hphoton2phin,r"#gamma_{2}")
leg3.Draw()

mcanvas.Draw()
```
